```python
def simplify(list1):
  ans=[]
  for i in list1:
    if(i>500000000 and i< 600000000):
      ans.append(55)
    elif(i>600000000 and i< 700000000):
      ans.append(65)
    elif(i>700000000 and i< 800000000):
      ans.append(75)
    elif(i>800000000 and i< 900000000):
      ans.append(85)
    elif(i>900000000 and i< 1000000000):
      ans.append(95)
    elif(i>1000000000 and i< 1100000000):
      ans.append(105)
    elif(i>1100000000 and i< 1200000000):
      ans.append(115)
    elif(i>1200000000 and i< 1300000000):
      ans.append(125)
    elif(i>1300000000 and i< 1400000000):
      ans.append(135)
    elif(i>1400000000 and i< 1500000000):
      ans.append(145)
  return ans
years = simplify(final_table["Population"])
final_table["Population"] = years
final_table
```

```python
def categorize_continents(data):
  asia = 0
  africa= 0
  europe= 0
  north_america = 0
  oceania =0
  south_america =0
  for i, j in zip(data["Continent"],data["2023"]):
      if i == "Asia":
        asia += j
      elif i == "Europe":
        europe += j
      elif i == "Africa":
        africa+= j
      elif i== "North America":
        north_america += j
      elif i== "Oceania":
        oceania += j
      elif i=="South America":
        south_america += j
  return [asia, africa, europe, north_america, south_america, oceania]
```
